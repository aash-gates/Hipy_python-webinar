FizzBuzz is a classic introductory coding challenge is often used by potential employers to filter prospective candidates for coding jobs. Let's give it a go.
Part 1

Loop through the numbers 1 to 50. Every time a multiple of 3 is encountered print "Fizz". Every time a multiple of 5 is encountered print "Buzz". If a number is a multiple of both 3 and 5 print "Buzz".


```python
# loop through 1 to 50
# print "Fizz" if divisible by 3, "Buzz" if divisible by 5 and "FizzBuzz" if divisible by both.

for i in range(50):
  print(i)
```

```python
# loop through 1 to 50
# print "Fizz" if divisible by 3, "Buzz" if divisible by 5 and "FizzBuzz" if divisible by both.

for i in range(1, 51):
  if i% 3 == 0 :
    print(i,"Fizz")
  if i%5 == 0 :
    print(i,"buzz")
```

Part 2

Re-work your solution to Part 1 above to create a function in Python that returns the "FizzBuzz" result for any integer.


```python
def fizz_buzz_check(n) :
  if n% 3 == 0 :
    print(n,"Fizz")
  if n%5 == 0 :
    print(n,"buzz")
```

```python
fizz_buzz_check(15)
```

```python
def fizz_buzz_check(n) :
  if n%3 == 0 and n%5 ==0 :
    return "FizzBuzz"
  elif n % 3 == 0 :
    return "Fizz"
  elif n % 5 ==0 :
    return "Buzz"
  return False
```

```python
fizz_buzz_check(7)
```

```python
with open("random_ints.txt" ,"r") as f:
  data = f.read()

data_list = data.split("\n")
print (data_list)

data_list_int = []
for item in data_list :
  data_list_int.append(int(item))

print(data_list_int) 
```
