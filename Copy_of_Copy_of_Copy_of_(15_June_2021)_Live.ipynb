## Part 1

Loop through the numbers 1 to 50. Every time a multiple of 3 is encountered print "Fizz". Every time a multiple of 5 is encountered print "Buzz". If a number is a multiple of both 3 and 5 print "Buzz".

```
# loop through 1 to 50
# print "Fizz" if divisible by 3, "Buzz" if divisible by 5 and "FizzBuzz" if divisible by both.

for i in range(1, 51):
  if i% 3 == 0 :
    print(i,"Fizz")
  if i%5 == 0 :
    print(i,"buzz")
   
```

```
import pandas as pd
#df = pd.read_excel("report.xlsx")

df = pd.read_csv("https://raw.githubusercontent.com/rtreharne/python_for_researchers/master/report.csv", error_bad_lines=False)
```

```
import pycountry as pyc

countries = set(df["Home Country"])

def get_country_name(code):
  try:
    return pyc.countries.get(alpha_2=code).name
  except:
    return "NA"

# use function above to create a corresponding list of country names (from codes)
country_names = [get_country_name(x) for x in df["Home Country"].tolist()]

# add country_name column to dataframe
df["country_name"] = country_names

grouped = df.groupby(by="country_name").count() # count number of registrations from each country

other = grouped[grouped["Home Country"] < 5] # create another dataframe containing countries that have less than 5 registrations

new = pd.DataFrame({"country_name": ["Other"], "Home Country": len(other)}) # create row to add to grouped frame
new = new.set_index("country_name")

visible_countries = grouped[grouped["Home Country"] >=5]
visible_countries = visible_countries.append(new)

fig, ax = plt.subplots(figsize=(7,7))

visible_countries.plot.pie(y="Home Country", ax=ax, legend=False, rotatelabels=270)
ax.set_ylabel("")
```
